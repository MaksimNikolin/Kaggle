**Import**

```python
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
df_train = pd.read_csv("/kaggle/input/mobile-price-classification/train.csv")
df_test = pd.read_csv("/kaggle/input/mobile-price-classification/test.csv")
```

**EDA**

```python
X_train = df_train.drop(['clock_speed', 'm_dep', 'n_cores', 'price_range'], axis = 1)
y_train = df_train.price_range
X_test = df_test.drop(['id', 'clock_speed', 'm_dep', 'n_cores'], axis = 1)
```

**Model**

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

parametr = {'n_estimators':[10, 20, 30], 'max_depth':[10, 15, 20], 'min_samples_leaf':range(2,5), 'min_samples_split':range(2,5)}
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
```

```python
import time
start = time.time()

np.random.seed(0)
gridsearchCV_clf = GridSearchCV(rf, parametr, cv=5)
gridsearchCV_clf.fit(X_train, y_train)
print(gridsearchCV_clf.best_params_)
print('-' * 45)

end = time.time()
print(f'it takes {end - start} seconds')
```

```python
best_clf = gridsearchCV_clf.best_estimator_
best_clf.score(X_train, y_train)
```

```python
predict = rf.predict(X_test)
best_clf.score(X_test, predict)
```

**Export**

```python
id_col = df_test['id']
submission_df = pd.DataFrame({"id": id_col, "target": predict})
```
